```python
from statsmodels.formula.api import ols
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

**Linear Regression model**

```python
X=df1.drop(['price'],axis=1)
Y=df1['price']
import statsmodels.api as sm
Xc = sm.add_constant(X)
```

```python
base = sm.OLS(Y, Xc).fit()
base.summary()
```

```python
lr_pred=base.predict(Xc)
```

```python
round(mean_squared_error(Y,lr_pred),2)
```

```python
round(np.sqrt(mean_squared_error(Y,lr_pred)),2)
```

From the above model:
* R-square=45%
* MSE=5.04
* RMSE=2.24

### Assumptions Of Linear Regression

* The dependent variable must be numeric
* Linear relationshipbetween dependent and independent variables
* Predictors must not show multicollinearity
* Independence of observationsshould exist (Absence of Autocorrelation)
* The error terms should be homoscedastic
* The error terms must follow normal distribution

**Checking of multicollinearity**

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif
```

```python
Variance_Inf=[vif(Xc.values,i) for i in range(Xc.shape[1])]
```

```python
VIF=pd.DataFrame()
VIF['features']=Xc.columns
VIF['variance_inf']=Variance_Inf
```

```python
VIF
```

*There is no presence of multicollinearity*

**Checking of Normality of Residuals**

```python
Residuals=base.resid
```

```python
sns.distplot(Residuals)
```

```python
import scipy.stats as stats
stats.probplot(Residuals, plot=plt)
plt.show()
```

```python
print(stats.jarque_bera(Residuals))
```

*Since p-values(0.0) is less than 0.05, we will conclude that residual is not normal*

```python
lny = np.log(Y)

model = sm.OLS(lny, Xc).fit()
model.summary()
```

```python
resids=model.resid
```

```python
print(stats.jarque_bera(resids))
```

*Since p-values(0.0) is less than 0.05, we will conclude that residual is not normal*

**Test for Homoscedasticity**

```python
import statsmodels.stats.api as sms
sms.het_goldfeldquandt(y=base.resid, x=Xc)
```

*Since p-value(0.9887625858985992) is greater than 0.05 we can fail to reject null hypothesis and conclude that the variance is homoscedasticity*

**Autocorrelation**

```python
import statsmodels.tsa.api as smt
smt.graphics.plot_acf(Residuals, lags=20)
plt.show()
```

*There is presence of auto-correlation*

**Linearity Of Relationship**

```python
import statsmodels.api as sm
sm.stats.diagnostic.linear_rainbow(base)
```

*Since the p-value(0.99999997851232) is greater than 0.05 we fail to reject null hypothesis.We conclude that there is linear Relationship between the variables*
